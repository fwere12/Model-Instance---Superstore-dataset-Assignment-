Transform categorical variables into dummies using the function **get_dummies()** (20 pts)
    * Focus on City, State, Region, EmployeeID, Segment, Category, Sub-Category, and Ship Mode.
    * The transformation of each categorical variable is worth *2.5 pts*

```python
# Specify the categorical columns you want to transform
categorical_columns = ['City', 'State', 'Region', 'EmployeeID', 'Segment', 'Category', 'Sub-Category', 'Ship Mode']

# Create an empty DataFrame to store the dummy variables
dummy_df = pd.DataFrame()

# Apply get_dummies() to each categorical column and concatenate the dummy variables
for column in categorical_columns:
    # Create dummy variables for the current column
    dummies = pd.get_dummies(superstoreData[column], prefix=column, drop_first=True)
    
    # Concatenate the dummy variables to the empty DataFrame
    dummy_df = pd.concat([dummy_df, dummies], axis=1)

# Concatenate the original DataFrame 'superstoreData' with the dummy variables 'dummy_df'
superstoreData = pd.concat([superstoreData, dummy_df], axis=1)

# Display the first few rows of the updated DataFrame to check the added columns
print(superstoreData.head())

# Alternatively, you can display the list of columns in the DataFrame
print(superstoreData.columns)
```

* Create a column called "Shipping Time" representing the number of days between the Order Date and the Ship Date (2.5 pts)
    * The difference between two *datetime64* type variables returns a *timedelta* type variable
    * You can use **dt.days** to convert a *timedelta* to the number of days. Also, you can use Google search to see how to convert a *timedelta* to an *int*. Here is an example https://www.kite.com/python/answers/how-to-convert-a-timedelta-column-in-a-pandas-dataframe-to-a-days-column-in-python

```python
# Convert "Order Date" and "Ship Date" to datetime if not already
superstoreData['Order Date'] = pd.to_datetime(superstoreData['Order Date'])
superstoreData['Ship Date'] = pd.to_datetime(superstoreData['Ship Date'])

# Create the "Shipping Time" column
superstoreData['Shipping Time'] = (superstoreData['Ship Date'] - superstoreData['Order Date']).dt.days
```

* Using **patsy** and **statsmodels**, fit four models that predict *Sales* (25 pts)
    * For each model, choose a different combination of predictors. You can include categorical variables (e.g., State, Region, etc.), and/or numerical variables (e.g., Shipping time, Price, etc.)
    * Each model should have at least 3 predictors
    * Each model is worth *5 pts*
    * Select the best model based on the R-Squared coefficient (5 pts)
        * The best model is the one that has the highest R-Squared coefficient

```python
import patsy
import statsmodels.api as sm

# Model 1: Predictors - 'Shipping Time', 'Price', 'Category'
formula1 = 'Sales ~ Q("Shipping Time") + Price + Category'
y1, X1 = patsy.dmatrices(formula1, data=superstoreData, return_type='dataframe')
model1 = sm.OLS(y1, X1).fit()

# Model 2: Predictors - 'Region', 'Quantity', 'Shipping Time'
formula2 = 'Sales ~ Region + Quantity + Q("Shipping Time")'
y2, X2 = patsy.dmatrices(formula2, data=superstoreData, return_type='dataframe')
model2 = sm.OLS(y2, X2).fit()

# Model 3: Predictors - 'State', 'Category', 'Price'
formula3 = 'Sales ~ State + Category + Price'
y3, X3 = patsy.dmatrices(formula3, data=superstoreData, return_type='dataframe')
model3 = sm.OLS(y3, X3).fit()

# Model 4: Predictors 'Segment', 'Shipping Time', 'Quantity'
formula4 = 'Sales ~ Segment + Q("Shipping Time") + Quantity'
y4, X4 = patsy.dmatrices(formula4, data=superstoreData, return_type='dataframe')
model4 = sm.OLS(y4, X4).fit()

# Compare R-squared values to select the best model
r_squared_values = [model1.rsquared, model2.rsquared, model3.rsquared, model4.rsquared]
best_model_index = r_squared_values.index(max(r_squared_values))

# Display the R-squared values and the best model
print("R-squared values:", r_squared_values)
print("Best model:", best_model_index + 1)  # Adding 1 to match the model numbering

# Printing our models
print("\nModel 1 Summary:\n", model1.summary())
print("\nModel 2 Summary:\n", model2.summary())
print("\nModel 3 Summary:\n", model3.summary())
print("\nModel 4 Summary:\n", model4.summary())
```

Note: The best model based on R-Squared coefficient is the third model that had the following predictors: 'State', 'Category', 'Price'

* Use the LogisticRegression model of scikit-learn to create a model instance (15 pts)
    * Create a list of predictors. You can include the dummies created earlier in the list of predictors (5 pts)
        * The list of predictors should have at least 5 variables
    * Following the lecture notes, fit the model and calculate the y_predict (5 pts)
    * Following the lecture notes, calculate the y_true and the accuracy percentage (5 pts)

```python
train = superstoreData[:5000]
test = superstoreData[5000:]
```

```python
# Creating 'IsProfitable' variable for the training set
train['IsProfitable'] = (train['Profit'] >= 0).astype(int)

# Creating 'IsProfitable' variable for the testing set
test['IsProfitable'] = (test['Profit'] >= 0).astype(int)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# List of predictors
predictors = ['Discount', 'Price', 'Shipping Time', 'Sub-Category_Paper', 'Ship Mode_Standard Class', 'City_Abilene', 'State_Arizona', 'EmployeeID_2','Segment_Corporate','Category_Office Supplies']

# Creating the Logistic Regression model
logreg_model = LogisticRegression()

# Fitting the model to the training data
logreg_model.fit(train[predictors], train['IsProfitable'])

# Calculating y_predict on the testing data
y_predict = logreg_model.predict(test[predictors])

# Calculating y_true (actual outcomes) on the testing data
y_true = test['IsProfitable']

# Calculating accuracy percentage
accuracy_percentage = accuracy_score(y_true, y_predict) * 100

# Displaying the accuracy percentage
print(f"Accuracy Percentage: {accuracy_percentage:.2f}%")
```
